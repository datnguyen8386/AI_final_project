```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# ==== 1. Chuáº©n bá»‹ dá»¯ liá»‡u ====
train_dir = '/content/drive/MyDrive/Project_AI/train'
test_dir = '/content/drive/MyDrive/Project_AI/test'

# TÄƒng cÆ°á»ng dá»¯ liá»‡u cho táº­p train
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,         # Xoay Â±20 Ä‘á»™
    width_shift_range=0.2,     # Dá»‹ch ngang
    height_shift_range=0.2,    # Dá»‹ch dá»c
    shear_range=0.2,           # Biáº¿n dáº¡ng cáº¯t nghiÃªng
    zoom_range=0.2,            # PhÃ³ng to / thu nhá»
    brightness_range=[0.7, 1.3], # Thay Ä‘á»•i Ä‘á»™ sÃ¡ng
    horizontal_flip=True,      # Láº­t ngang
    fill_mode='nearest'        # Äiá»n pháº§n trá»‘ng khi xoay
)

# Test giá»¯ nguyÃªn, chá»‰ chuáº©n hÃ³a 0â€“1
test_datagen = ImageDataGenerator(rescale=1./255)

train_data = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical'
)

test_data = test_datagen.flow_from_directory(
    test_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical'
)

# ==== 2. XÃ¢y dá»±ng mÃ´ hÃ¬nh CNN ====
model = Sequential([
    Conv2D(32, (3,3), activation='relu', padding='same', input_shape=(128,128,3)),
    Conv2D(32, (3,3), activation='relu', padding='same'),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu', padding='same'),
    Conv2D(64, (3,3), activation='relu', padding='same'),
    MaxPooling2D(2,2),

    Conv2D(128, (3,3), activation='relu', padding='same'),
    Conv2D(128, (3,3), activation='relu', padding='same'),
    MaxPooling2D(2,2),
    Conv2D(256, (3,3), activation='relu', padding='same'),
    Conv2D(256, (3,3), activation='relu', padding='same'),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(train_data.num_classes, activation='softmax')
])

# ==== 3. Compile mÃ´ hÃ¬nh ====
model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# ==== 4. Huáº¥n luyá»‡n ====
history = model.fit(
    train_data,
    epochs=100,                # tÄƒng sá»‘ epoch náº¿u cÃ³ nhiá»u dá»¯ liá»‡u
    validation_data=test_data
)

```

```python
# =============================================================================
# ğŸ± SMART FOOD TRAY v21.0 - CHUáº¨N KHAY INOX 6 NGÄ‚N (Final)
# =============================================================================

import os
import sys

# --- 1. CÃ€I Äáº¶T NHANH ---
print("ğŸš€ Äang thiáº¿t láº­p há»‡ thá»‘ng nháº­n diá»‡n khay 6 ngÄƒn... (Chá» 10s)")
os.system("pip install -q gradio tensorflow opencv-python matplotlib plotly pandas pillow")

import gradio as gr
import cv2
import numpy as np
import pandas as pd
from PIL import Image, ImageOps
import tensorflow as tf
import plotly.graph_objects as go
from google.colab import drive

# --- 2. Káº¾T Ná»I DRIVE & MODEL ---
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive', force_remount=False)

MODEL_PATH = '/content/drive/MyDrive/project_foodAI.h5'

# Danh sÃ¡ch mÃ³n (Cá»§a báº¡n)
MON_AN_LABELS = ['Canh chua cÃ³ cÃ¡', 'Canh chua khÃ´ng cÃ³ cÃ¡', 'Canh rau', 'CÃ¡ hÃº kho', 'CÆ¡m tráº¯ng', 'Rau xÃ o', 'SÆ°á»n nÆ°á»›ng', 'Thá»‹t kho', 'Thá»‹t kho trá»©ng', 'Trá»©ng chiÃªn', 'Äáº­u há»§ sá»‘t cÃ ']
CALO_TABLE = {'Canh chua cÃ³ cÃ¡': 150, 'Canh chua khÃ´ng cÃ³ cÃ¡': 90, 'Canh rau': 70, 'CÃ¡ hÃº kho': 250, 'CÆ¡m tráº¯ng': 130, 'Rau xÃ o': 120, 'SÆ°á»n nÆ°á»›ng': 350, 'Thá»‹t kho': 300, 'Thá»‹t kho trá»©ng': 340, 'Trá»©ng chiÃªn': 200, 'Äáº­u há»§ sá»‘t cÃ ': 180}
GIA_TABLE = {'Canh chua cÃ³ cÃ¡': 25000, 'Canh chua khÃ´ng cÃ³ cÃ¡': 10000, 'Canh rau': 7000, 'CÃ¡ hÃº kho': 30000, 'CÆ¡m tráº¯ng': 10000, 'Rau xÃ o': 10000, 'SÆ°á»n nÆ°á»›ng': 30000, 'Thá»‹t kho': 25000, 'Thá»‹t kho trá»©ng': 30000, 'Trá»©ng chiÃªn': 25000, 'Äáº­u há»§ sá»‘t cÃ ': 25000}

try:
    classifier_model = tf.keras.models.load_model(MODEL_PATH)
    MODEL_READY = True
    print("âœ… ÄÃ£ load Model thÃ nh cÃ´ng.")
except:
    classifier_model = None
    MODEL_READY = False
    print("âš ï¸ Cháº¡y cháº¿ Ä‘á»™ Demo.")

# --- 3. THUáº¬T TOÃN Cáº®T 6 Ã” (TINH CHá»ˆNH CHO áº¢NH Cá»¦A Báº N) ---
def crop_6_slots_v2(image_pil):
    target_w, target_h = 800, 600
    img = image_pil.resize((target_w, target_h))
    img_np = np.array(img)
    h, w = img_np.shape[:2]

    # Tá»ŒA Äá»˜ CHUáº¨N CHO KHAY Cá»¦A Báº N (ÄÃ£ cÄƒn chá»‰nh)
    # Format: (y_start%, y_end%, x_start%, x_end%)
    slots_coords = [
        # HÃ ng trÃªn (2 ngÄƒn lá»›n)
        {"id": 1, "pos": (0.06, 0.48, 0.04, 0.43)},  # TrÃ¡i TrÃªn (CÆ¡m)
        {"id": 2, "pos": (0.06, 0.48, 0.56, 0.96)},  # Pháº£i TrÃªn (Canh)

        # Khe giá»¯a (ÄÅ©a muá»—ng) -> Bá» QUA
        {"id": 6, "pos": (0.10, 0.90, 0.45, 0.54), "is_tool": True},

        # HÃ ng dÆ°á»›i (3 ngÄƒn nhá»)
        {"id": 3, "pos": (0.55, 0.94, 0.04, 0.32)}, # TrÃ¡i DÆ°á»›i (Máº·n)
        {"id": 4, "pos": (0.55, 0.94, 0.36, 0.63)}, # Giá»¯a DÆ°á»›i (XÃ o)
        {"id": 5, "pos": (0.55, 0.94, 0.67, 0.96)}, # Pháº£i DÆ°á»›i (Trá»©ng/Máº·n)
    ]

    crops = []
    overlay = img_np.copy()

    for slot in slots_coords:
        y1, y2, x1, x2 = slot["pos"]
        py1, py2 = int(y1*h), int(y2*h)
        px1, px2 = int(x1*w), int(x2*w)

        crop_img = img_np[py1:py2, px1:px2]

        # Váº½ khung: MÃ u xÃ¡m cho khe Ä‘Å©a, MÃ u xanh cho mÃ³n Äƒn
        is_tool = slot.get("is_tool", False)
        color = (150, 150, 150) if is_tool else (0, 255, 0)
        thickness = 2 if is_tool else 4

        cv2.rectangle(overlay, (px1, py1), (px2, py2), color, thickness)

        # Ghi nhÃ£n
        label_text = "Dua/Muong" if is_tool else f"Mon {slot['id']}"
        cv2.putText(overlay, label_text, (px1+10, py1+30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)

        crops.append({
            "img": crop_img,
            "is_tool": is_tool,
            "id": slot["id"]
        })

    return Image.fromarray(overlay), crops

# --- 4. Xá»¬ LÃ ---
def process_tray(image_input):
    if image_input is None: return None, [], pd.DataFrame(), "âš ï¸ ChÆ°a cÃ³ áº£nh", None

    # Chuáº©n hÃ³a áº£nh
    img_pil = image_input.convert("RGB")
    img_pil = ImageOps.exif_transpose(img_pil)

    # Cáº¯t áº£nh
    mask_vis, crops_data = crop_6_slots_v2(img_pil)

    total_calo = 0
    total_price = 0
    gallery_data = []
    table_data = []

    for item in crops_data:
        # Náº¿u lÃ  khay Ä‘Å©a muá»—ng -> Bá» qua
        if item["is_tool"]:
            continue

        # Nháº­n diá»‡n
        crop_img = item["img"]
        if MODEL_READY:
            try:
                c_rz = cv2.resize(crop_img, (128, 128))
                img_array = np.expand_dims(c_rz.astype("float32") / 255.0, axis=0)
                preds = classifier_model.predict(img_array, verbose=0)[0]
                idx = int(np.argmax(preds))
                label = MON_AN_LABELS[idx]
            except: label = "Lá»—i"
        else:
            label = MON_AN_LABELS[item["id"] % len(MON_AN_LABELS)] # Demo

        calo = CALO_TABLE.get(label, 0)
        price = GIA_TABLE.get(label, 0)

        total_calo += calo
        total_price += price

        gallery_data.append((Image.fromarray(crop_img), label))
        table_data.append({"Vá»‹ trÃ­": f"Ã” sá»‘ {item['id']}", "MÃ³n": label, "Kcal": calo, "GiÃ¡": f"{price:,}"})

    # Dashboard
    summary_html = f"""
    <div style="background: white; padding: 20px; border-radius: 15px; box-shadow: 0 4px 15px rgba(0,0,0,0.1); text-align: center;">
        <div style="font-size: 14px; color: #888; margin-bottom: 5px;">Káº¾T QUáº¢ PHÃ‚N TÃCH</div>
        <div style="display: flex; justify-content: space-around; align-items: center;">
            <div>
                <div style="font-size: 32px; font-weight: 800; color: #e74c3c;">{total_calo}</div>
                <div style="font-size: 12px; font-weight: 600; color: #888;">KCAL</div>
            </div>
            <div style="width: 1px; height: 40px; background: #eee;"></div>
            <div>
<div style="font-size: 32px; font-weight: 800; color: #27ae60;">{total_price//1000}k</div>
                <div style="font-size: 12px; font-weight: 600; color: #888;">VNÄ</div>
            </div>
        </div>
        <div style="margin-top: 15px; font-size: 13px; color: #2980b9; font-weight: 500;">
            âœ… ÄÃ£ quÃ©t 5 Ã´ thá»©c Äƒn (Bá» qua khay Ä‘Å©a)
        </div>
    </div>
    """

    fig = go.Figure(data=[go.Pie(labels=[d["MÃ³n"] for d in table_data], values=[d["Kcal"] for d in table_data], hole=.6)])
    fig.update_layout(showlegend=False, margin=dict(t=10, b=10, l=10, r=10), height=200)

    return mask_vis, gallery_data, pd.DataFrame(table_data), summary_html, fig

# --- 5. GIAO DIá»†N ---
css = """
body { font-family: sans-serif; background: #f4f4f9; }
.container { max-width: 1000px; margin: auto; background: white; padding: 20px; border-radius: 20px; box-shadow: 0 10px 30px rgba(0,0,0,0.05); }
.btn-run { background: #27ae60 !important; color: white !important; font-weight: bold; border-radius: 10px; height: 50px; }
"""

with gr.Blocks(css=css, theme=gr.themes.Soft(), title="SmartTray Fixed") as demo:
    with gr.Column(elem_classes="container"):
        gr.Markdown("## ğŸ± Há»‡ thá»‘ng nháº­n diá»‡n Khay CÆ¡m 6 NgÄƒn")

        with gr.Row():
            with gr.Column(scale=4):
                input_img = gr.Image(sources=["upload", "webcam"], type="pil", height=300, label="áº¢nh Ä‘áº§u vÃ o")
                btn = gr.Button("ğŸš€ PHÃ‚N TÃCH NGAY", elem_classes="btn-run")

            with gr.Column(scale=6):
                summary = gr.HTML("Káº¿t quáº£ sáº½ hiá»‡n á»Ÿ Ä‘Ã¢y...")
                chart = gr.Plot(label="Biá»ƒu Ä‘á»“")

        with gr.Row():
            with gr.Column():
                mask_out = gr.Image(label="Vá»‹ trÃ­ Ä‘Ã£ cáº¯t", height=300)
            with gr.Column():
                gallery = gr.Gallery(label="Chi tiáº¿t mÃ³n", columns=3, height=300)

        table = gr.Dataframe(label="Báº£ng chi tiáº¿t")

    btn.click(process_tray, inputs=[input_img], outputs=[mask_out, gallery, table, summary, chart])

print("ğŸš€ ÄÃ£ sáºµn sÃ ng! HÃ£y upload áº£nh khay cÆ¡m 6 ngÄƒn cá»§a báº¡n...")
demo.queue().launch(share=True, debug=False)
```
